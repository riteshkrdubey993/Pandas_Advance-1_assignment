Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that
iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The
function should print the sum to the console.
For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should
calculate and print the sum of the first three values, which is 60.

```python
def sum_of_three(data,n): #take two arguments, Data an number n = total sum upto "n"
    df = pd.DataFrame(data, columns = ['values'])
    calculate = df[0:n]
    total = calculate.sum()
    
    return print('The sum of the first three values is : ',total )

data = np.random.randint(5,100,size = 10)
sum_of_three(data,3)

print(data)
```

Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column
'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
data = {'Text': ["Hello, how are you?", "I'm doing well, thanks.", "What about you?",
                 "I'm a language model.", "Nice to meet you!"]}

def word_count(Data):
    df = pd.DataFrame(Data)
    df['Word_Count'] = df['Text'].apply(lambda text: len(text.split()))
    
    return df

word_count(data)
```

Q5. How are DataFrame.size() and DataFrame.shape() different?

DataFrame.size:
1. This attribute returns the total number of elements in the DataFrame.
2. It calculates the size by multiplying the number of rows (DataFrame.shape[0]) by the number of columns (DataFrame.shape[1]).
3. It gives you the total count of cells in the DataFrame, including all cells, regardless of their content or data type.

```python
data = {'A': [10, 20, 30, 40],
        'B': [100, 200, 300, 400],
        'C': [1000, 2000, 3000, 4000]}

df = pd.DataFrame(data)
df.size
```

DataFrame.shape:
1. This attribute returns a tuple representing the dimensions of the DataFrame.
2. It returns a tuple (n_rows, n_columns), where n_rows is the number of rows in the DataFrame, and n_columns is the number of columns.
3. It gives you an understanding of the DataFrame's structure in terms of rows and columns.

```python
data = {'A': [10, 20, 30, 40],
        'B': [100, 200, 300, 400],
        'C': [1000, 2000, 3000, 4000]}

df = pd.DataFrame(data)
df.shape
```

Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email
addresses in the format 'username@domain.com'. Write a Python function that creates a new column
'Username' in df that contains only the username part of each email address.
The username is the part of the email address that appears before the '@' symbol. For example, if the
email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your
function should extract the username from each email address and store it in the new 'Username'
column.

```python

def user_name(data):
    df = pd.DataFrame(data)
    df['username'] = df['Email'].apply(lambda user: user.split('@')[0])
    
    return df

data = {'Email': ['user1@example.com',
                  'user2@example.com',
                  'user3@example.com',
                  'user4@example.com',
                  'user5@example.com',
                  'user6@example.com',
                  'user7@example.com',
                  'user8@example.com',
                  'user9@example.com',
                  'user10@example.com',
                  'user11@example.com',
                  'user12@example.com',
                  'user13@example.com',
                  'user14@example.com',
                  'user15@example.com',
                  'user16@example.com']}


user_name(data)
```

Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,
median, and standard deviation of the values in the 'Values' column.

```python
def stats_value(DataPoints):
    df = pd.DataFrame(DataPoints,columns = ['values'])
    
    mean = df['values'].mean()
    median = df['values'].median()
    mode = df['values'].mode()
    
    print('Mean of values is: ', mean)
    print('Median of values is: ', median)
    print('Mode of values is:', mode)
   
    
data_points = np.random.randint(low = 1, high = 50, size = 1000)
stats_value(data_points)
```
